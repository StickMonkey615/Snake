```python
!pip install pygame #instal pygame. Is this needed every run?
import pygame
import time
import random

pygame.init()

#define display boundaries
x_min = 0
x_max = 800
y_min = 0
y_max = 600

dis=pygame.display.set_mode((x_max,y_max)) #initialise screen dimensions
pygame.display.update()
pygame.display.set_caption('Snake')

#define colours
white=(255,255,255)
blue=(0,0,255)
red=(255,0,0)
black=(0,0,0)

snake_speed = 15
block_size = 10

#create a function to build up snake from multiple blocks defined in a list
def our_snake(block_size, snake_List):
  for x in snake_List:
    pygame.draw.rect(dis, black, [x[0], x[1], block_size, block_size])

#create a function to display a message
####

#create a function for the looping game stages
def game_loop():
  
  x1 = x_max/2 #set centre screen point as start point
  y1 = y_max/2

  x1_change = 0
  y1_change = 0

  game_over = False

  snake_List = []
  snake_size = 1

  foodx = round(random.randrange(0, x_max - block_size) / 10.0) * 10.0
  foody = round(random.randrange(0, y_max - block_size) / 10.0) * 10.0
  print(foodx)

  clock = pygame.time.Clock() #create clock object

  while not game_over: #loop to keep screen open until game is complete
    for event in pygame.event.get():
      if event.type==pygame.QUIT:
        game_over=True
      if event.type == pygame.KEYDOWN:
        if event.key == pygame.K_LEFT:
          x1_change = -block_size
          y1_change = 0
        elif event.key == pygame.K_RIGHT:
          x1_change = block_size
          y1_change = 0
        elif event.key == pygame.K_UP:
          x1_change = 0
          y1_change = -block_size
        elif event.key == pygame.K_DOWN:
          x1_change = 0
          y1_change = block_size

    #check if edge of screen reached
    if x1 < x_min or x1 > x_max or y1 < y_min or y1 > y_max:
      #display message that game over
      #change snake colour to red
      pygame.display.update()
      game_over = True

    x1 += x1_change
    y1 += y1_change
    dis.fill(white)
    pygame.draw.rect(dis, blue, [foodx, foody, block_size,block_size]) #create food
    snake_Head = []
    snake_Head.append(x1)
    snake_Head.append(y1)
    snake_List.append(snake_Head)
    
    if len(snake_List) > snake_size: #remove oldest enty in list
      del snake_List[0]

    for x in snake_List[:-1]:
      if x == snake_Head:
        game_over = True
        print('You ate yourself!')

    our_snake(block_size, snake_List)

    if x1 == foodx and y1 ==foody:
      print('Yummy!') #print yummy message
      snake_size += 1 #increase length of snake
      foodx = round(random.randrange(0,x_max - block_size) / 10.0) * 10.0
      foody = round(random.randrange(0,y_max - block_size) / 10.0) * 10.0
      #pygame.draw.rect(dis, black, [x1,y1,block_size,block_size]) #redraw snake longer
    pygame.display.update()

    clock.tick(snake_speed)      

  pygame.quit()
  quit()
game_loop()
```
