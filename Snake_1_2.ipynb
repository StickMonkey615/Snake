```python
import pygame
import time
import random

pygame.init()

#define display boundaries
x_min = 0
x_max = 600
y_min = 0
y_max = 400

dis=pygame.display.set_mode((x_max,y_max)) #initialise screen dimensions
pygame.display.update()
pygame.display.set_caption('Snake')

#define colours
white=(255,255,255)
blue=(0,0,255)
red=(255,0,0)
black=(0,0,0)
green = (0, 255, 0)

block_size = 10

#a function to build up snake from multiple blocks defined in a list
def our_snake(block_size, snake_List):
  for x in snake_List:
    pygame.draw.rect(dis, black, [x[0], x[1], block_size, block_size])

#a function to draw multiple food entities
def draw_food(blovk_size, food_List):
  for x in food_List:
    pygame.draw.rect(dis,green, [x[0], x[1], block_size, block_size])

#a function to display a message
def message(text, colour):
  msg = pygame.font.SysFont("bahnschrift", 15).render(text, True, colour)
  dis.blit(msg,[x_max / 6,y_max / 3])

#a function for displaying current score
def current_score(score):
  msg = pygame.font.SysFont("comicsansms", 25).render("Your score = " + str(score), True, green)
  dis.blit(msg,[0, 0])

#a function for the looping game stages
def game_loop():
  
  x1 = x_max/2 #set centre screen point as start point
  y1 = y_max/2

  x1_change = 0
  y1_change = 0
  last_direction = 0 #used to prevent changes of direction back on self ending game 0=init, 1=L, 2=R, 3=U, 4=D

  game_over = False
  game_end = False

  snake_List = []
  snake_size = 1
  snake_speed = 10

  foodx = round(random.randrange(0, x_max - block_size) / 10.0) * 10.0
  foody = round(random.randrange(0, y_max - block_size) / 10.0) * 10.0
  food_List = []
  food_Point = []
  food_Point.append(foodx)
  food_Point.append(foody)
  food_List.append(food_Point)

  clock = pygame.time.Clock() #create clock object

  while not game_over: #loop to keep screen open until game is complete

    while game_end == True:
      message("GAME OVER! Press P to play again, or Q to quit", red)
      pygame.display.update()

      for event in pygame.event.get():
        if event.type == pygame.KEYDOWN:
          if event.key == pygame.K_q:
            game_end = False
            game_over = True
          if event.key == pygame.K_p:
            game_loop()

    for event in pygame.event.get(): #loop to check for arrow key presses or window closed
      if event.type==pygame.QUIT:
        game_over=True
      if event.type == pygame.KEYDOWN:
        if event.key == pygame.K_LEFT and last_direction != 2:
          x1_change = -block_size
          y1_change = 0
          last_direction = 1
        elif event.key == pygame.K_RIGHT and last_direction != 1:
          x1_change = block_size
          y1_change = 0
          last_direction = 2
        elif event.key == pygame.K_UP and last_direction != 4:
          x1_change = 0
          y1_change = -block_size
          last_direction = 3
        elif event.key == pygame.K_DOWN and last_direction != 3:
          x1_change = 0
          y1_change = block_size
          last_direction = 4

    #check if edge of screen reached
    if x1 < x_min or x1 > x_max-block_size or y1 < y_min or y1 > y_max-block_size:
      #display message that game over
      #change snake colour to red
      pygame.display.update()
      game_end = True

    #check if game has begun
    if x1_change == 0 and y1_change == 0:
      message("Press an arrow key to begin", black)

    #calculate moved position of snake elements
    x1 += x1_change
    y1 += y1_change
    dis.fill(blue)
    #pygame.draw.rect(dis, green, [foodx, foody, block_size,block_size]) #create food
    
    snake_Head = []
    snake_Head.append(x1)
    snake_Head.append(y1)
    snake_List.append(snake_Head)
    if len(snake_List) > snake_size: #remove oldest enty in list
      del snake_List[0]

    #check if snake has collided with self
    for x in snake_List[:-1]:
      if x == snake_Head:
        game_end = True

    our_snake(block_size, snake_List) #redraw snake in moved position
    draw_food(block_size, food_List)
    current_score(snake_size - 1)

    #check if food eaten
    if x1 == foodx and y1 ==foody:
      snake_size += 1 #increase length of snake
      snake_speed += 1
      del food_List[0]
      #generate new food location
      foodx = round(random.randrange(0,x_max - block_size) / 10.0) * 10.0
      foody = round(random.randrange(0,y_max - block_size) / 10.0) * 10.0
      food_Point = []
      food_Point.append(foodx)
      food_Point.append(foody)
      food_List.append(food_Point)
    
    pygame.display.update()

    clock.tick(snake_speed)

  pygame.quit()
  quit()
game_loop()
```
